```python
## Creating the function
def greeting_someone():
  print("Hello, Good Morning")
  print("It's a beautiful day.")

## Calling the Function
greeting_someone()
```

```python
## adding arguements in the functions
## arguements are some values which we pass to the function and the value goes inside the function.

def greeting_someone(name): ## since one arguement is there, we need to pass a value while calling the func
  print(f"Hello {name} !! , Good Morning")
  print("It's a beautiful day.")

## Calling the Function
greeting_someone("Anik")
greeting_someone("Carol")
greeting_someone("Jhon")
```

```python
## Write a function to check wheather a particular int is odd pr even

def odd_even_checker(num):
  if num%2==0:
    print(f"The number {num} is even")
  else:
    print(f"The number {num} is odd")

## Calling the Function
odd_even_checker(10)
odd_even_checker(11)
odd_even_checker(99)
odd_even_checker(100)
```

```python
## Write a function to add two integers.
def add(num1, num2):
  result = num1 + num2
  print(f"Result : {result}")

## Calling the Function
add(10,15)
add(9, -4)
add(5, 6)
add(6, -5)
add(-5, 5)
```
